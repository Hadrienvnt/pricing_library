```python
from src.asset_class import EuropeanOption, OptionType
from src.market_data import YieldCurve, YieldCurveMethod
from src.black_scholes import BlackScholesPricer
```

# 1. Constructing the yield curve

```python
# Invented data for illustration purpose
market_tenors = [0.1, 0.5, 1.0, 2.0, 3.0, 5.0, 7.0, 10.0, 15.0, 20.0, 30.0]
market_rates =  [0.010, 0.012, 0.015, 0.018, 0.020, 0.025, 0.028, 0.030, 0.032, 0.033, 0.034]
```

```python
# Yield Curve:
yc = YieldCurve(market_tenors, market_rates, method=YieldCurveMethod.NELSON_SIEGEL)
yc.plot(title="Yield Curve (Nelson-Siegel)")
```

# 2. Option Pricing

```python
#  OPTION SETUP
K = 100.0
T = 2.0
my_option = EuropeanOption(strike=K, expiry=T, option_type=OptionType.CALL)
print(f"Option Created: Call K={K}, T={T}")
```

```python
# Pricing:
print("Pricing Using Black-Scholes:")  
volatility = 0.20
bs_engine = BlackScholesPricer(volatility=volatility, risk_free_rate=yc)
my_option.set_pricing_method(bs_engine)
my_option.plot_analysis()
```
