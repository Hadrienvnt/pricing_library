```python
# Libraries
from src.asset_class import OptionType, AsianOption, BarrierType, BarrierOption
from src.monte_carlo import AsianOptionMCPricer,BarrierOptionMCPricer
```

```python
# Options & Market Data
K = 100
T = 0.5             # 6 months 
r = 0.03            # flat curve
sigma = 0.20
option_type = OptionType.CALL
```

# 1. Asian Option Pricing

Let's analyse a Asian call with strike K=100 and maturity T=6mo:

```python
# Option Setup & Pricing
asian_option = AsianOption(strike=K,expiry=T,option_type=option_type)

asian_pricer = AsianOptionMCPricer(volatility=sigma,risk_free_rate=r)
asian_option.set_pricing_method(asian_pricer)

asian_option.plot_analysis(spot_min=80,spot_max=120)
```

# 2. Barrier Option Pricing

Let's analyse a Up&In Barrier Call with stike K=100, maturity T=6mo and barrier H=105

```python
# Option Setup & Pricing
barrier = 105
barrier_type = BarrierType.UP_AND_IN
barrier_option = BarrierOption(strike=K,expiry=T,option_type=option_type,barrier=barrier,barrier_type=barrier_type)


barrier_pricer = BarrierOptionMCPricer(volatility=sigma,risk_free_rate=r)
barrier_option.set_pricing_method(barrier_pricer)

barrier_option.plot_analysis(spot_min=80,spot_max=120)
```
